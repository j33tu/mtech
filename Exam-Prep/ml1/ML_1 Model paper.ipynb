```python
Q1 = df['enginesize'].quantile(0.25)
Q3 = df['enginesize'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

df1 = df[(df['enginesize'] >= lower_bound) & (df['enginesize'] <= upper_bound)]
```

```python
sns.boxplot(data=df1, y='enginesize')
```

```python
df1['car_vol'] = df1.carwidth * df1.carheight * df1.carlength
df1 = df1.drop(columns = ['carlength','carwidth','carheight'])
```

```python
df1['avg_mpg'] = (df1.citympg + df1.highwaympg) / 2
df1 = df1.drop(columns = ['citympg','highwaympg'])
```
