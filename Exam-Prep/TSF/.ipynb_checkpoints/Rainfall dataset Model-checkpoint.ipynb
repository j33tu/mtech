# Rainfall Dataset for Simple Time Series Analysis

This dataset contains daily rainfall measurements (in millimeters) for the year 2022. The data spans from January 1, 2022, to July 3, 2022, covering a total of 184 days. The dataset can be used for various machine learning tasks, such as time series forecasting, pattern recognition, or anomaly detection related to rainfall patterns.

Column Descriptors:

date (date):
Description: The date of the rainfall measurement in the format YYYY-MM-DD.

Example: 2022-01-01

rainfall (float):
Description: The amount of rainfall recorded on the corresponding date, measured in millimeters (mm).

Example: 12.5

Range: The rainfall values range from 0.0 mm (no rainfall) to 22.4 mm (the maximum recorded value in the dataset).

### a. Decompose the Time Series

* Multiplicative seasonality is not appropriate for zero and negative values

```python
# Decompose the time series
decomposition = seasonal_decompose(df['rainfall'], model='additive', period=30)  # Assuming monthly seasonality

# Plot the decomposition
plt.figure(figsize=(12, 8))
decomposition.plot()
plt.show()
```

Inference:

1. Trend: The smooth line with an upward trajectory indicates a long-term increase in rainfall over the period.
2. Seasonal: The consistent, repeating patterns suggest a regular seasonal effect on rainfall.
3. Residual: The random scatter of points suggests that after removing the trend and seasonal effects, the remaining variations are irregular and do not follow a systematic pattern.

This decomposition helps in understanding the underlying structure of the time series and in building more accurate forecasting models by accounting for these components separately.

### b. Perform Dickey-Fuller Test to Check Stationarity

```python
from statsmodels.tsa.stattools import adfuller

# Perform the Dickey-Fuller test
adf_test = adfuller(df['rainfall'])
print('ADF Statistic:', adf_test[0])
print('p-value:', adf_test[1])

# Check for stationarity
if adf_test[1] > 0.05:
    print("The series is non-stationary.")
else:
    print("The series is stationary.")
```

### c. Plot Auto Correlation and Partial Auto Correlation Function

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF 
plt.figure(figsize=(10, 6))

plot_acf(df['rainfall'].dropna(), lags=30, ax=plt.gca())

plt.title('Autocorrelation Function')

plt.show()
```

```python
# Plot  PACF
plt.figure(figsize=(10, 6))

plot_pacf(df['rainfall'].dropna(), lags=30, ax=plt.gca())
plt.title('Partial Autocorrelation Function')

plt.show()
```

* Lag Selection: The significant spikes in the graph, which are outside the confidence interval (shaded area), suggest the lags at which the partial autocorrelation is significant.
* Model Order: Typically, the number of significant spikes indicates the order of the AR(p) model; in this case, it appears there are a few lags with significant partial autocorrelation.
* Statistical Significance: The lags where the blue dots fall outside the shaded area are considered statistically significant and should be included in the AR model.

### a. Fit a Base Model and Observe the Residuals, RMSE, and MAPE Values

#### Fit a base ARIMA model and calculate metrics

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

# Fit a base ARIMA model
model = ARIMA(train['rainfall'], order=(3, 0, 3))  # Simple ARIMA model with p=3, d=0, q=3
model_fit = model.fit()
print(model_fit.summary())
```

```python
# Forecast the test set
forecast = model_fit.forecast(steps=len(test))
test['forecast'] = forecast
```

```python
# Calculate RMSE and MAPE
rmse = np.sqrt(mean_squared_error(test['rainfall'], test['forecast']))
mape = mean_absolute_percentage_error(test['rainfall'], test['forecast'])

print(f'RMSE: {rmse}')
print(f'MAPE: {mape}')
```

Root Mean Squared Error (RMSE): Approximately 2.57

Mean Absolute Percentage Error (MAPE): Approximately 1.69 x 10^15

Inference:

* The RMSE indicates that, on average, the models predictions are off by approximately 2.57 units of rainfall.
* The extremely high MAPE suggests that the models predictions are not accurate and vary significantly from the actual values.
Given these metrics, further investigation and potential model improvement are necessary to enhance accuracy. Consider exploring alternative models or refining the existing ARIMA configuration.

```python
# Plot the original data and the forecast
plt.figure(figsize=(10, 6))
plt.plot(test['rainfall'], label='Actual Data')
plt.plot(test.index, forecast, label='Forecast', color='red')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('ARIMA Forecast')
plt.legend()
plt.show()
```

```python
# Plot residuals
residuals = model_fit.resid
plt.figure(figsize=(10, 6))
plt.plot(residuals,color='red')
plt.title('Residuals of the ARIMA Model')
plt.show()
```

```python
# Plot residuals distribution
plt.figure(figsize=(10, 6))
plt.hist(residuals, bins=30)
plt.title('Histogram of Residuals')
plt.show()
```

### b. Improve the Model

Optimize SARIMA Parameters: 
We will use a grid search to find the best (p, d, q) and seasonal (P, D, Q, S) parameters.

#### Grid search for the best ARIMA parameters

```python
# Perform grid search for ARIMA hyperparameters (optional step for better model)
import itertools
import warnings
warnings.filterwarnings("ignore")

p = d = q = range(0, 4)
pdq = list(itertools.product(p, d, q))
best_aic = float("inf")
best_pdq = None

for param in pdq:
    try:
        temp_model = ARIMA(train['rainfall'], order=param)
        temp_model_fit = temp_model.fit()
        if temp_model_fit.aic < best_aic:
            best_aic = temp_model_fit.aic
            best_pdq = param
    except:
        continue

print(f'Best ARIMA parameters: {best_pdq} with AIC: {best_aic}')
```

### Using Seasonal Differencing and SARIMA:

Optimize SARIMA Parameters: We will use auto_arima to find the best SARIMA parameters (P, D, Q, S).

```python
from pmdarima import auto_arima

# Use auto_arima to find the best SARIMA parameters
smodel = auto_arima(df['rainfall'], seasonal=True, m=30, 
                    stepwise=True, suppress_warnings=True, 
                    trace=True)

print(smodel.summary())
```

The AIC value of 429.744 is used to compare models; a lower AIC indicates a better fit. The selected model has a relatively low AIC, suggesting it fits the data well compared to other models.

###  Refit the model and make predictions:

```python
# Fit the best model from auto_arima
best_model = smodel.fit(train['rainfall'])

# Make predictions
final_test_predictions = best_model.predict(n_periods=len(test))
final_test_predictions = pd.Series(final_test_predictions, index=test.index)
```

```python
# Calculate RMSE and MAPE for final model
final_rmse = mean_squared_error(test['rainfall'], final_test_predictions, squared=False)
final_mape = mean_absolute_percentage_error(test['rainfall'], final_test_predictions)

print(f'Final RMSE: {final_rmse}')
print(f'Final MAPE: {final_mape}')
```

* RMSE (Root Mean Squared Error):
An RMSE of 0.6888 indicates that the model's predictions deviate from the actual values by approximately 0.69 mm of rainfall on average. This low RMSE suggests that the model's predictions are fairly accurate in terms of absolute error.

* MAPE (Mean Absolute Percentage Error):
A MAPE of 52240125820183.21 is extremely high, indicating that the model's percentage error in predictions is very large. This suggests that while the absolute errors (as indicated by RMSE) are low, the relative errors are significantly high, possibly due to very small actual rainfall values in the dataset.

# Further improvement of the model using log1p transformation

The RMSE for the final model has improved, but the MAPE is still extremely high, indicating potential issues with the data or model. MAPE values that are extremely high often suggest that there are outliers or zero values in the data which are causing issues.

Further Improvement and Diagnostics
To further investigate and refine the model, we should:

1. Check for Zero or Near-Zero Values: These can cause issues with MAPE calculation.
2. Analyze the Distribution of Rainfall Values: Determine if there are significant outliers.
3. Consider Log Transformation: If there are no zero values, a log transformation might stabilize variance.
4. Check Residual Diagnostics: Ensure residuals are normally distributed and have constant variance.

```python
# Check for zero or near-zero values
zero_values = df[df['rainfall'] == 0]
print(f'Number of zero or near-zero values: {len(zero_values)}')
```

```python
# Plot the distribution of rainfall values
plt.figure(figsize=(10, 6))
plt.hist(df['rainfall'], bins=30)
plt.title('Distribution of Rainfall Values')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Apply log1p transformation
endog_log = np.log1p(endog)
```

```python
# Fit the SARIMAX model on the log-transformed data
model_fit_log = sm.tsa.statespace.SARIMAX(endog_log, order=(2, 0, 0), seasonal_order=(0, 1, 0, 30)).fit()

# Calculate residuals on the log-transformed model
final_residuals_log = model_fit_log.resid

# Ensure residuals are a numpy array
final_residuals_log = np.array(final_residuals_log)

# Plot the residuals
plt.figure(figsize=(10, 6))
plt.plot(final_residuals_log, color='red')
plt.title('Residuals of Log-Transformed Final Model')
plt.xlabel('Date')
plt.ylabel('Residuals')
plt.show()
```

```python
# ACF and PACF of residuals
fig, ax = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(final_residuals_log, ax=ax[0])
plot_pacf(final_residuals_log, ax=ax[1])
plt.show()

# Ljung-Box test
ljung_box_results_log = sm.stats.acorr_ljungbox(final_residuals_log, lags=[10], return_df=True)
print("Ljung-Box test results:\n", ljung_box_results_log)

# Histogram and Q-Q plot of residuals
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
sns.histplot(final_residuals_log, kde=True, ax=ax[0])
ax[0].set_title('Histogram of Residuals')

sm.qqplot(final_residuals_log, line='s', ax=ax[1])
ax[1].set_title('Q-Q Plot of Residuals')
plt.show()
```

```python
# Forecast the next 30 days using the log-transformed model
forecast_steps = 30
forecast_log = model_fit_log.get_forecast(steps=forecast_steps)

# Extract forecasted values and confidence intervals
forecast_values_log = forecast_log.predicted_mean
confidence_intervals_log = forecast_log.conf_int()

# Inverse transform the forecasted values and confidence intervals
forecast_values = np.expm1(forecast_values_log)
confidence_intervals = np.expm1(confidence_intervals_log)
```

```python
# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(endog, label='Observed')
plt.plot(forecast_values, label='Forecast', color='red')
plt.fill_between(confidence_intervals.index,
                 confidence_intervals.iloc[:, 0],
                 confidence_intervals.iloc[:, 1], color='pink', alpha=0.3)
plt.title('Births Forecast for the Next 30 Days (Log-Transformed Model)')
plt.legend()
plt.show()
```
