### 4b.ii)Conduct a hypothesis to see whether there is a difference in terms of salary for students degree specialization of management and engineering. What do you conclude (consider level of significance 0.05) (5 marks)

````python
from scipy.stats import ttest_ind

# Filter data for students with degree specialization in management
management_salary = df[df['Course_Degree'] == 'Management']['Salary']

# Filter data for students with degree specialization in engineering
engineering_salary = df[df['Course_Degree'] == 'Engineering']['Salary']

# Perform independent samples t-test
t_statistic, p_value = ttest_ind(management_salary, engineering_salary)

# Print results
print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in mean salary between students with degree specializations in management and engineering.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in mean salary between students with degree specializations in management and engineering.")
````

### 4b.iii)Conduct an hypothesis to prove whether the salary of a student from commenrce stream is higher than Science stream, do you agree or not. (consider level of significance 0.05)  (5 marks)

````python
from scipy.stats import ttest_ind

# Filter data for students from the commerce stream
commerce_salary = df[df['Stream_HSC'] == 'Commerce']['Salary']

# Filter data for students from the science stream
science_salary = df[df['Stream_HSC'] == 'Science']['Salary']

# Perform independent samples t-test
t_statistic, p_value = ttest_ind(commerce_salary, science_salary)

# Print results
print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The mean salary of students from the commerce stream is significantly higher than that of students from the science stream.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in mean salary between students from the commerce stream and students from the science stream.")
````
