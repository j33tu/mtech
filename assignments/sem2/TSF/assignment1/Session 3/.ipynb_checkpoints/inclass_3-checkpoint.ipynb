```python
import pandas                          as      pd
import numpy                           as      np
import matplotlib.pyplot               as      plt
import seaborn as sns
from   IPython.display                 import  display
from   pylab                           import  rcParams 
from   datetime                        import  datetime, timedelta
from statsmodels.tsa.stattools         import  adfuller,acf,pacf
from statsmodels.graphics.tsaplots     import  plot_pacf,plot_acf
from statsmodels.graphics.gofplots     import qqplot
from statsmodels.tsa.seasonal          import seasonal_decompose
from statsmodels.tsa.arima.model       import ARIMA
import warnings
warnings.filterwarnings('ignore')
```

## read AirTemp data and convert it to time series

```python
df = pd.read_csv("AirTemp.csv")
df.head()
```

```python
df['Month'].unique()
```

```python
def year_month_to_date(year,month):
    month_dict = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4,
                'May': 5, 'Jun': 6, 'Jul': 7, 'Aug': 8,
                'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
                }
    mnt = month_dict[month]
    return pd.to_datetime(f"{year}-{mnt:02d}-01")
```

```python
df['Date'] = df.apply(lambda x : year_month_to_date(x['Year'],x['Month']), axis=1)
df.head()
```

```python
df.dtypes
```

```python
df.drop(columns=['Year','Month'],inplace=True)
df.set_index('Date',inplace=True)
df.head()
```

## plot the time series

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df,x=df.index,y='AvgTemp',label='AvgTemp')
plt.show()
```

# Check the stationarity of series using ADF test

```python
observations= df.values
test_result = adfuller(observations)
test_result
```

```python
print('ADF Statistic: %f' % test_result[0])
print('p-value: %f' % test_result[1])
print('Critical Values:')
for key, value in test_result[4].items():
	print('\t%s: %.5f' % (key, value))
```

### test results confirms that the series is stationary

```python
decomp = seasonal_decompose(df,model='add')
decomp.plot()
plt.plot()
```

# Build ARMA model for selected p and q

```python
#as per acf & pacf --> q=2 and p=1

model_ARIMA = ARIMA(train,order = (1,0,2)) # (p,d,q)
model_ARIMA_fit = model_ARIMA.fit()
model_ARIMA_fit.summary()
```

# Find predictions of model for the range of test data

```python
forecast_ARIMA = model_ARIMA_fit.forecast(steps=len(test))

plt.figure(figsize=(12,5))
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.plot(forecast_ARIMA, label='ARIMA forecats')
plt.legend(loc='best')
plt.title("ARIMA(1,0,2)")
plt.grid()
```

# Find residuals for the model and plot using Q-Q- plot
