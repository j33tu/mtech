```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
df  = pd.read_csv('Lung_Feat.csv')
df.head()
```

```python
features = list(df.columns)
features.remove('T')
```

```python
sns.pairplot(df)
```

```python
scaler = StandardScaler()
df_std = scaler.fit_transform(df[features])
```

```python
num_components = 20  # You can choose the number of components you want.
pca = PCA(n_components=num_components)
```

```python
pca
```

```python
pca_result = pca.fit_transform(df_std)
```

```python
pca_df = pd.DataFrame(data=pca_result, columns=[f'PC{i}' for i in range(1, num_components + 1)])
pca_df.head()
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
cumulative_explained_variance
```

```python
n_components = np.argmax(cumulative_explained_variance >= 0.95) + 1
n_components
```

```python
# 10 features give 95% of the variance
```

```python
plt.figure(figsize=(8, 6))
plt.bar(range(1, num_components + 1), explained_variance_ratio * 100, alpha=0.7, align='center')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance (%)')
plt.title('Explained Variance Ratio for Each Principal Component')
plt.xticks(range(1, num_components + 1))
plt.show()
```

```python
pca.components_[0]
```

```python
top_features = []
for component in range(10):
    component_weights = pca.components_[component]
    top_feature_indices = np.argsort(np.abs(component_weights))[::-1][:5]
    top_features.append(top_feature_indices)
```

```python
top_features
```

```python
# Create a mapping from column indices to column names
column_names = df.columns

# List the top feature names for the first five principal components
top_feature_names = []
for feature_indices in top_features:
    top_feature_names.append([column_names[i] for i in feature_indices])

# Print or display the top feature names
for component, features in enumerate(top_feature_names, 1):
    print(f"Top Features for Principal Component {component}: {features}")
```

```python
# Create a dictionary to store the cumulative absolute weights for each feature
feature_weights = {}

# Loop through the top features for each principal component
for feature_indices in top_features:
    for idx in feature_indices:
        # Calculate the absolute weight for the feature
        absolute_weight = np.abs(pca.components_[:, idx]).sum()
        
        # Update the cumulative weight for the feature
        if idx in feature_weights:
            feature_weights[idx] += absolute_weight
        else:
            feature_weights[idx] = absolute_weight

# Sort the features by cumulative absolute weight in descending order
sorted_feature_weights = {k: v for k, v in sorted(feature_weights.items(), key=lambda item: item[1], reverse=True)}

# Get the top 5 features with the highest cumulative absolute weights
top_influential_features = list(sorted_feature_weights.keys())[:5]

# Print or display the top 5 most influential features
print("Top 5 Most Influential Features:")
for idx in top_influential_features:
    print(f"Feature {idx}: {column_names[idx]}")
```
