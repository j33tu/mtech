```python
##define matrix A B C 
A = np.array([[5,6,2],
              [4,7,1],
              [0,3,1]
])
B = np.array([[1,-2,1],
              [4,4,5],
              [5,5,1]
])
C = np.array([[1,0,1],
              [4,1,0],
              [2,0,1]
])
```

```python
C
```

```python
#(AB)^t - BA traspose

left=np.dot(A,B)
```

```python
np.transpose(left)
```

```python
np.dot(np.transpose(B),np.transpose(A)) == np.transpose(left)
```

```python
np.array_equal(np.dot(np.transpose(B),np.transpose(A)),np.transpose(left))
```

```python
dot=np.dot(A,B)
np.linalg.inv(dot)
```

```python
np.dot(np.linalg.inv(B),np.linalg.inv(A))
```

```python
np.allclose(np.linalg.inv(dot),np.dot(np.linalg.inv(B),np.linalg.inv(A)))
```
