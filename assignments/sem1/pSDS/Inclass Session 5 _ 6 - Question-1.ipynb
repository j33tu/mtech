**Note:** A bubble chart is a type of scatter plot that represents three dimensional data. The third variable is represented by the size of a point (marker).

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
x = np.random.rand(40)# Random x-coordinates
y = np.random.rand(40)# Random y-coordinates
sizes = np.random.randint(100, 800, size=50)  # Random bubble sizes
colors = np.random.rand(4)  # Random bubble colors

# Create the bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=sizes, c=colors, alpha=0.7, cmap='viridis')

# Add colorbar for better visualization of the 'colors' parameter
cbar = plt.colorbar()
cbar.set_label('Color Value')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bubble Plot with Adjusted Transparency (alpha)')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'diamonds' dataset from Seaborn
diamonds = sns.load_dataset('diamonds')

# Multiple Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='cut', y='price', data=diamonds)
plt.xlabel('Cut')
plt.ylabel('Price')
plt.title('Multiple Boxplot for Price based on Diamond Cut')
plt.show()

# Violin Plot
plt.figure(figsize=(8, 6))
sns.violinplot(x='cut', y='price', data=diamonds)
plt.xlabel('Cut')
plt.ylabel('Price')
plt.title('Violin Plot for Price based on Diamond Cut')
plt.show()

# Boxen Plot
plt.figure(figsize=(8, 6))
sns.boxenplot(x='cut', y='price', data=diamonds)
plt.xlabel('Cut')
plt.ylabel('Price')
plt.title('Boxen Plot for Price based on Diamond Cut')
plt.show()

#error about 'cut'
```
