```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 6, 9, 12, 15]
y3 = [1, 3, 5, 7, 9]

# Create the multiple line plot
plt.figure(figsize=(8, 6))

# Line plot for y1
plt.plot(x, y1, label='Line 1', marker='o', linestyle='-', linewidth=2)

# Line plot for y2
plt.plot(x, y2, label='Line 2', marker='s', linestyle='--', linewidth=2)

# Line plot for y3
plt.plot(x, y3, label='Line 3', marker='^', linestyle='-.', linewidth=2)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot')
plt.legend()

plt.grid(True)  # Add grid lines

plt.show()
```
