```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x1 = [1,2,3,4,5,6]
y1 = [10,13,14,15,16,17]
plt.scatter(x=x1,y=y1,color='green',marker='>') #bivariant needs two values 

plt.xlabel('x-asix')
plt.ylabel('y-asix')
plt.title('my first plot')
plt.xticks([1,2,3])
plt.xticks([10,15,20])
plt.savefig('scatter.png')
f=plt.gcf() get current figure object 
print(f)
plt.show()
```

```python

import numpy as np 

x1 = np.array([1,2,3,4,5,6])
y1 = x1 * 2
y2 = x1 ** 2
plt.plot(x1,y1,color='green',marker='*',label='linear plot')
plt.plot(x1,y2,color='red',marker='>',label='expo plot')
#plt.show()
print(y2)
```

```python
df1.sort_values(by='wt').plot(kind='line',x='wt',y='mpg',marker='*',figsize=(15,5))
df1.sort_values(by='wt').plot(kind='line',x='wt',y='disp',marker='*',figsize=(15,5))
```

```python
plt.figure(figsize=(15,5),dpi=100)
plt.plot(df1.sort_values(by='wt').wt,df1.disp,color='green',marker='*',label='linear plot')
plt.plot(df1.sort_values('wt').wt,df1.mpg,color='red',marker='>',label='expo plot')
plt.legend()
plt.show()

```
