```python
q1=data['age'].quantile(.25)
q3=data['age'].quantile(.75)
IQR=q3-q1
ub=q3+1.5*IQR
lb=q1-1.5*IQR

data= data[~((data['age']<lb) | (data['age']>ub))]
```

```python
data.skew()
```

```python
sns.pairplot(data)
```

```python
y=data['strength']
X=data.drop('strength',axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
model1=lr.fit(x_train,y_train)
ypred=model1.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error

r2=r2_score(y_test,ypred)
rmse=np.sqrt(mean_squared_error(y_test,ypred))

print(r2)
print(rmse)
```

```python
# feature selection
# mlxtend

from mlxtend.feature_selection import SequentialFeatureSelector as sfs
```

```python
# forward feature selector
lr= LinearRegression()
sfs_for= sfs(estimator=lr, k_features='best' ,forward=True)
model2= sfs_for.fit(x_train,y_train)
```

```python
model2.estimator
```

```python
sfs_forward=list(model2.k_feature_names_)
```

```python
# Backward elimination
lr= LinearRegression()
sfs_back= sfs(estimator=lr, k_features='best' ,forward=False)
model3= sfs_back.fit(x_train,y_train)
```

```python
sfs_backward=list(model3.k_feature_names_)
```

```python
model3.scoring
```

```python
model3.k_score_
```

```python
print(sfs_backward)
print(sfs_forward)
```

```python
from sklearn.feature_selection import RFE


rfe= RFE(estimator=lr,n_features_to_select=6)
model4=rfe.fit(x_train,y_train)
```

```python
model4.ranking_
```

```python
rank=pd.DataFrame()
rank['feat']=X.columns
rank['rank']=model4.ranking_
best=rank[rank['rank']==1]
```

```python
rfe_feat=list(best['feat'])
```

```python
rfe_feat
```
