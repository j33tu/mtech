# Pivot table

```python
import pandas as pd
df1=pd.read_csv('https://raw.githubusercontent.com/vcsgithub/vcsgithub.github.io/master/lds1.csv')
df1
dfp = pd.pivot_table(data=df1,index='Sector',aggfunc='mean',values=['Exp','Rev'])
dfp
```

```python
dfp = pd.pivot_table(data=df1,index='Sector',columns='GST',values=['Exp','Rev'],aggfunc='count')
dfp
```

```python
dfp = pd.pivot_table(data=df1,index='Sector',columns='GST',values=['Exp','Rev'],aggfunc='mean')
dfp
dfp.loc['Pub',('Rev','Services')]
```

```python
#across each sector i want to see mean max and diff between both
dfp=pd.pivot_table(df1,index='Sector',values='Exp',aggfunc=['max',np.min,lambda ps2:ps2.max()-ps2.min()])
dfp
```

```python

def sales_performance_cat(sales_values):
    if sales_values > 950:
        return 'excellent sales'
    elif sales_values < 930:
        return 'average sales'
    else:
        return 'good sales'
sales_performance_cat(94)
```
