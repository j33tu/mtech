Blurring (Smoothing):

Gaussian Blur: Applies a Gaussian kernel to smooth the image. The cv2.GaussianBlur function takes the image, kernel size, and standard deviation as arguments.


Sharpening:

Filter2D: Applies a custom kernel to sharpen the image. The kernel used here is designed to enhance edges.


Edge Detection:

Canny Edge Detection: Detects edges in an image using gradient values. The cv2.Canny function takes the grayscale image and two threshold values as arguments.


Histogram Equalization:

EqualizeHist: Enhances the contrast of the grayscale image by spreading out the intensity values. The cv2.equalizeHist function takes the grayscale image as an argument.

```python
print("Shape of the image: ", image_rgb.shape)
print("Data type of the image: ", image_rgb.dtype)
print("Size of the image: ", image_rgb.size)
```

```python
# Apply Gaussian Blur
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

# Convert to RGB for display
blurred_image_rgb = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)

# Display the blurred image
plt.imshow(blurred_image_rgb)
plt.title('Gaussian Blur')
plt.axis('off')
plt.show()
```

```python
# Define a sharpening kernel
import numpy as np
sharpening_kernel = np.array([[-1, -1, -1],
                              [-1,  9, -1],
                              [-1, -1, -1]])

# Apply the sharpening kernel to the image
sharpened_image = cv2.filter2D(image, -1, sharpening_kernel)

# Convert to RGB for display
sharpened_image_rgb = cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB)

# Display the sharpened image
plt.imshow(sharpened_image_rgb)
plt.title('Sharpened Image')
plt.axis('off')
plt.show()
```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Canny edge detection
edges = cv2.Canny(gray_image, 100, 200)

# Display the edges
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection (Canny)')
plt.axis('off')
plt.show()
```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(gray_image)

# Display the original grayscale and equalized images
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Histogram Equalized Image')
plt.axis('off')
plt.show()
```

# Image enhancement and filtering


```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Add Gaussian noise to the image
def add_gaussian_noise(image, mean=0, std=25):
    gauss = np.random.normal(mean, std, image.shape).astype('uint8')
    noisy_image = cv2.add(image, gauss)
    return noisy_image

noisy_image = add_gaussian_noise(image_rgb)

# Convert the noisy image to grayscale
gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

# Apply Median Blur
median_blur = cv2.medianBlur(gray_image, 5)

# Apply Histogram Equalization
equalized_image = cv2.equalizeHist(median_blur)

# Display the images
plt.figure(figsize=(20, 5))

plt.subplot(1, 5, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 5, 2)
plt.imshow(noisy_image)
plt.title('Noisy Image')
plt.axis('off')

plt.subplot(1, 5, 3)
plt.imshow(median_blur, cmap='gray')
plt.title('Median Blurred Image')
plt.axis('off')

plt.subplot(1, 5, 4)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
