```python
import ipaddress

# Original CIDR block
original_cidr = ipaddress.ip_network('10.64.0.0/10')

# Calculate the /16 subnet
subnet_16 = list(original_cidr.subnets(new_prefix=16))[0]

# Calculate the /21 subnet
# Ensuring the /21 subnet does not overlap with the /16 subnet
# We take the next available /21 subnet after the /16 range
subnet_21_start = list(subnet_16.hosts())[-1] + 1
subnet_21 = ipaddress.ip_network(f'{subnet_21_start}/21', strict=False)

# Calculate the /27 subnet
# Ensuring the /27 subnet does not overlap with the /21 subnet
# We take the next available /27 subnet after the /21 range
subnet_27_start = list(subnet_21.hosts())[-1] + 1
subnet_27 = ipaddress.ip_network(f'{subnet_27_start}/27', strict=False)

# Print the new subnets
print(f"Subnet /16: {subnet_16}")
print(f"Subnet /21: {subnet_21}")
print(f"Subnet /27: {subnet_27}")
```
