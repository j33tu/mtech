```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv('CC_DATA.csv')
df.head()
```

```python
df.drop('CUST_ID',inplace=True,axis=1)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].mean(), inplace=True)
```

```python
df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].mean(), inplace=True)

df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
inp_sc=sc.fit_transform(df)
from sklearn.decomposition import PCA
pc=PCA()
```

```python
pca_input=pc.fit_transform(inp_sc)
pca_input.shape
```

```python
pd.DataFrame(pca_input).head(1)
```

```python
#eign values 
pd.DataFrame(pc.explained_variance_).round(2)
```

```python
pd.DataFrame(pc.explained_variance_)[0:1]/pd.DataFrame(pc.explained_variance_).sum()
```

```python
pd.DataFrame(pc.explained_variance_ratio_)
```

```python
cum_var=np.cumsum(pc.explained_variance_ratio_)
(cum_var<0.95).sum()
```
