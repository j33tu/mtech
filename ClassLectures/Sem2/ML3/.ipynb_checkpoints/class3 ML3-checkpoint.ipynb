```python
import numpy as np

import pandas as pd
```

```python
df=pd.read_csv('BC.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
input=df.drop(['id','diagnosis'],axis=1)
```

```python
out=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(input,out,test_size=0.3,random_state=48)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
rf=RandomForestClassifier(random_state=0)
rf.fit(xtrain,ytrain)
```

```python
ypred=rf.predict(xtest)

print('with out any dimentation reduction')
print("number of shape:",input.shape[1])
print(classification_report(ytest,ypred))
```

```python
# scal down your data before applying pca or svd
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
inp_sc=sc.fit_transform(input)
```

```python
from sklearn.decomposition import PCA
pc=PCA()
```

```python
pca_input=pc.fit_transform(inp_sc)
pca_input.shape,inp_sc.shape
```

```python
pd.DataFrame(inp_sc).head(1)
```

```python
pd.DataFrame(pca_input).head(3)
```

```python
#eign values 
pd.DataFrame(pc.explained_variance_).round(2)
```

```python
pd.DataFrame(pc.explained_variance_)[0:1]/pd.DataFrame(pc.explained_variance_).sum()
```

```python
pd.DataFrame(pc.explained_variance_ratio_)
```

```python
cum_var=np.cumsum(pc.explained_variance_ratio_)
(cum_var<0.95).sum()
```

```python
pca_input_reduced=pca_input[:,:9]
pca_input_reduced.shape
xtrain,xtest,ytrain,ytest=train_test_split(pca_input_reduced,out,test_size=0.3,random_state=48)
xtrain,xtest,ytrain,ytest=train_test_split(pca_input_reduced,out,test_size=0.3,random_state=48)
rf=RandomForestClassifier(random_state=0)
rf.fit(xtrain,ytrain)
ypred=rf.predict(xtest)

print('with out any dimentation reduction')
print("number of shape:",input.shape[1])
print(classification_report(ytest,ypred))
```

```python
import time

t0 = time.time()

from sklearn.decomposition import TruncatedSVD
sv=TruncatedSVD(n_components=29)
svd_inp=sv.fit_transform(inp_sc)
cum_var=np.cumsum(sv.explained_variance_ratio_)
(cum_var<0.95).sum()
svd_input_reduced=svd_inp[:,:9]
svd_input_reduced.shape
xtrain,xtest,ytrain,ytest=train_test_split(svd_input_reduced,out,test_size=0.3,random_state=48)
rf=RandomForestClassifier(random_state=0)
rf.fit(xtrain,ytrain)
ypred=rf.predict(xtest)

print('with out any dimentation reduction')
print("number of shape:",input.shape[1])
print(classification_report(ytest,ypred))

t1 = time.time()

total = t1-t0
print('time taken:',total)
```
