```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('db_data.csv')
```

```python
data.head(2)
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
plt.scatter(data.iloc[:,0],data.iloc[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
inp=data.drop('out',axis=1)
```

```python
clust_mod=KMeans(n_clusters=2,random_state=0,max_iter=1000)
clust_mod.fit(inp)
label=clust_mod.labels_
```

```python
plt.scatter(data.iloc[:,0],data.iloc[:,1],c=label)
```

```python
from sklearn.cluster import DBSCAN
db_mod=DBSCAN(eps=0.2,min_samples=5)
db_mod.fit(inp)
label_db=db_mod.labels_
plt.scatter(data.iloc[:,0],data.iloc[:,1],c=label_db)
```
