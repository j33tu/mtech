```python
#SIDE TO SIDE 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(3,2),dpi=100)
cnames=['tata','birla','wirpo','infy','hcl']
xticks=np.arange(len(cnames))
print(xticks)
#print(xticks)
e1count=[100,200,300,900,800]
e2count=[150,250,350,450,550]
plt.bar(x=xticks+.4,height=e1count,color='red',width=.4)
plt.bar(x=cname,height=e2count,color='green',width=.4)
plt.xticks(xticks+.2,cnames)
plt.show
```
